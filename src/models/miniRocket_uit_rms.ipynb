# Minirocket testing

```python
le = preprocessing.LabelEncoder()

X_train, y_train = load_from_tsfile_to_dataframe(module_path + '/features/extracted_features_ts_files/uit_MiniROCKET_TRAIN.ts')
X_test, y_test = load_from_tsfile_to_dataframe(module_path + '/features/extracted_features_ts_files/uit_MiniROCKET_TEST.ts')
#le.fit(y_train)
#y_train = le.transform(y_train)
#y_test = le.transform(y_test)

X_train = from_nested_to_2d_array(X_train)
X_test = from_nested_to_2d_array(X_test)

X_train.columns = np.arange(len(X_train.columns))
X_test.columns = np.arange(len(X_test.columns))


y_train = pd.Series(y_train)
y_test = pd.Series(y_test)


#k = 200
#select = SelectKBest(chi2, k=k)
#X_train_s = select.fit(X_train, y_train)
#indices = select.get_support(indices = True)
#X_test = X_test[indices]
#X_train = X_train[indices]

pca = PCA(n_components = 200)

pca.fit(X_train)
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

```python
pca = PCA(n_components = 69)
pca.fit(X_train)
exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)

plt.plot(exp_var_cumul)
```

```python
pls2 = PLSRegression(n_components=2)
p2 = pls2.fit_transform(X_train, pd.factorize(y_train)[0].astype(int))
f, ax = helper.scatterplot_with_colors(p2[0],pd.factorize(y_train)[0].astype(int),
                                       module_path, new_legends = list(pd.factorize(y_train)[1]),
                                       x_y_labels = ['PLSR1', 'PLSR2'])

plt.show()

#f.savefig(figure_path + f'PLSR_minirocket_lung.pdf', bbox_inches='tight')
```

```python
from sklearn.naive_bayes import GaussianNB

from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler() 
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train)) 
X_test = pd.DataFrame(scaler.transform(X_test)) 

clf = GaussianNB().fit(X_train, y_train)
y_pred = clf.predict(X_test)

f, ax = utility.plot_cm(y_test, y_pred, module_path = module_path, class_names = pd.factorize(y_train)[1])
```

```python
from sklearn.svm import SVC

clf = SVC(**{'C': 10, 'gamma': 1, 'kernel': 'rbf', 'probability' : True}).fit(X_train, y_train)

y_pred = clf.predict(X_test)
prob = clf.predict_proba(X_test)

f, ax = utility.plot_cm(y_test, y_pred, module_path = module_path, class_names = pd.factorize(y_train)[1])
```

```python
from sklearn.neighbors import KNeighborsClassifier


clf = KNeighborsClassifier(n_neighbors=4, weights='distance').fit(X_train, y_train)

y_pred = clf.predict(X_test)
prob = clf.predict_proba(X_test)

f, ax = utility.plot_cm(y_test, y_pred, module_path = module_path, class_names = pd.factorize(y_train)[1])
```
