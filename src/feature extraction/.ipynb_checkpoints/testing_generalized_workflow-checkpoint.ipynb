# Preproject method with Lung Sound (kaggle)

```python
ts_path = module_path + '/data/ts_files/crackleNoCrackle_FULL.ts'
#NB ! This file has already been downsampled and denoised before saving
kwargs_ls = {
    'ts_file_path': ts_path,
    'sr' : 8000,
    'UCR_file': False,
    'n_mfcc' : 30,
    'denoise' : False,
    'downsample' : False,
    'downsample_new_sr' : 8000,
    'update' : False,
    'name' : 'lungsound'
}

ls = preproject.overproduced_audio_time_series_features(**kwargs_ls)
```

```python
X_train, X_test, y_train, y_test = ls.get_X_y_transformed(train_test_split = True)
```

```python
pls2 = PLSRegression(n_components=2)
p2 = pls2.fit_transform(X_train, y_train)
f, ax = helper.scatterplot_with_colors(p2[0], y_train.values.astype(int),
                                       module_path, new_legends = ['no crackle', 'crackle'],
                                       x_y_labels = ['PLSR1', 'PLSR2'])

plt.show()

f.savefig(figure_path + f'PLSR_preproject_lung.pdf', bbox_inches='tight')
```

# Minirocket testing

```python
le = preprocessing.LabelEncoder()

X_train, y_train = load_from_tsfile_to_dataframe(module_path + '/features/extracted_features_ts_files/lungsound_MiniROCKET_TRAIN.ts')
X_test, y_test = load_from_tsfile_to_dataframe(module_path + '/features/extracted_features_ts_files/lungsound_MiniROCKET_TEST.ts')
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)

X_train = from_nested_to_2d_array(X_train)
X_test = from_nested_to_2d_array(X_test)

X_train.columns = np.arange(len(X_train.columns))
X_test.columns = np.arange(len(X_test.columns))


y_train = pd.Series(y_train)
y_test = pd.Series(y_test)

'''
k = 200
select = SelectKBest(chi2, k=k)
X_train_s = select.fit(X_train, y_train)
indices = select.get_support(indices = True)
X_test = X_test[indices]
X_train = X_train[indices]

'''
```

```python
pls2 = PLSRegression(n_components=2)
p2 = pls2.fit_transform(X_train, y_train)
f, ax = helper.scatterplot_with_colors(p2[0], y_train.values.astype(int),
                                       module_path, new_legends = ['no crackle', 'crackle'],
                                       x_y_labels = ['PLSR1', 'PLSR2'])

plt.show()

f.savefig(figure_path + f'PLSR_minirocket_lung.pdf', bbox_inches='tight')
```

### Dimentionality reduction using PCA

```python
pca = PCA(n_components = 100)
pca.fit(X_train)
exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)

plt.plot(exp_var_cumul)
```

```python
pca = PCA(n_components = 10)

pca.fit(X_train)
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

### Balancing the dataset

```python
y_train.value_counts()
```

```python
indices_0 = np.where(y_train.astype(int) == 0)[0]
indices_1 = np.where(y_train.astype(int) == 1)[0]
indices = np.concatenate((indices_0[0:1757], indices_1))
```

### Plotting the performance

Here a Gaussian Process classifier is utilized. This works well because we want a problibalistic classifier

```python
kernel = 1.0 * RBF(1.0)
gpc = GaussianProcessClassifier(kernel=kernel,random_state=0, n_jobs = -3)
gpc.fit(X_train[indices],y_train[indices])
pred = gpc.predict(X_test)

f, ax = utility.plot_cm(y_test, pred, module_path = module_path, class_names = ['no crackle', 'crackle'])

plt.show()

#f.savefig(figure_path + f'lung_minirocket_cm.pdf', bbox_inches='tight')
```
