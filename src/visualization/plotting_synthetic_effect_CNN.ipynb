# Plotting the f1 and accuracy score of different syntetic datasets

```python
import sys
import os
import pandas as pd
import numpy as np

module_path = os.path.abspath(os.path.join('../..'))

sys.path.insert(1, module_path + '/src')
import utility

sys.path.insert(1, module_path + '/src/models/')
import helper

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import matplotlib
import matplotlib.pyplot as plt

figure_path = module_path + '/figures/'
```

```python
import pickle

# Frequency Scores

with open(module_path + '/src/visualization/synthetic_increase_acc_f1/acc_freqs_5.pkl', 'rb') as f:
    acc_freq = pickle.load(f)
    
with open(module_path + '/src/visualization/synthetic_increase_acc_f1/f1_freqs_5.pkl', 'rb') as f:
    f1_freq = pickle.load(f)


# Amplitude Scores

with open(module_path + '/src/visualization/synthetic_increase_acc_f1/acc_ampl_5.pkl', 'rb') as f:
    acc_ampl = pickle.load(f)
    
with open(module_path + '/src/visualization/synthetic_increase_acc_f1/f1_ampl_5.pkl', 'rb') as f:
    f1_ampl = pickle.load(f)


# Duration scores 
with open(module_path + '/src/visualization/synthetic_increase_acc_f1/acc_dur_5.pkl', 'rb') as f:
    acc_dur = pickle.load(f)
    
with open(module_path + '/src/visualization/synthetic_increase_acc_f1/f1_dur_5.pkl', 'rb') as f:
    f1_dur = pickle.load(f)
    
    
freqs = np.linspace(100, 2000, 20).astype(int)
ampls = np.linspace(0.5, 10, 20)
durs = np.linspace(0.01, 0.2, 20)
```

```python
font = FontProperties(fname = module_path + '/src/visualization/CharterRegular.ttf', size = 10, weight = 1000)
font_small = FontProperties(fname = module_path + '/src/visualization/CharterRegular.ttf', size = 8, weight = 1000)
colors = ['#F94144', '#90BE6D', '#577590','#F3722C', '#F8961E', '#F9844A', '#F9C74F', '#43AA8B', '#4D908E', '#277DA1']
colors_opa = ['#F9414466', '#90BE6D66', '#57759066','#F3722C66', '#F8961E66',
              '#F9844A66', '#F9C74F66', '#43AA8B66', '#4D908E66', '#277DA166']

f, ax = plt.subplots(1,1)
f.set_size_inches(6, 2)


ax.plot(durs, acc_dur, c = colors[2], label = 'Accuracy', linewidth = 0.5)
ax.plot(durs, f1_dur, c = colors[0], label = 'F1 score', linewidth = 0.5)


ax.set_ylabel('Performance', fontproperties = font)
ax.set_xlabel('Duration of synthetic abnormality [s]', fontproperties = font)


for label in ax.get_yticklabels() :
    label.set_fontproperties(font_small)
for label in ax.get_xticklabels() :
    label.set_fontproperties(font_small)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(0.5)
    
ax.legend(prop = font)

plt.savefig(figure_path + 'duration_synthetic_increase_performance_5s.pdf', bbox_inches='tight')
```
